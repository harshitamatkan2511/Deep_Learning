```python
#importing packages
import pandas as pd
from sklearn.model_selection import train_test_split
import tensorflow as tf
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Input
```

```python
#Reading dataset
body_fat=pd.read_csv("bodyfat (1).csv")
```

```python
body_fat
```

```python
#separating into dependent and independent
x=body_fat.drop("BodyFat",axis=1)
y=body_fat["BodyFat"]
```

```python
#splitting into train and test
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,train_size=0.8,random_state=0)
```

```python
x.shape
```

```python
y.shape
```

```python
#creating the model with the hidden layers
model = Sequential([
    Input(shape=(14,),name='il',),
    Dense(128,activation='relu',name='d1'),
    Dense(64,activation='relu',name='d2'),
    Dense(32,activation='relu',name='d3'),
    Dense(1,activation='linear',name='ol')])
```

```python
model.summary()
```

```python
model.compile(loss = 'mean_squared_error', optimizer = Adam(learning_rate=0.1), metrics=['RootMeanSquaredError'])
```

```python
epochs=100
```

```python
#fitting to the model
model.fit(x_train,y_train,epochs=epochs)
```

```python
#evaluating the model
loss, accuracy = model.evaluate(x_test, y_test)
```

```python
print("Test Loss:", loss)
print("Test Rmse:", accuracy)
```
